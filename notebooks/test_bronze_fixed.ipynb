```python
# notebooks/test_bronze_final.ipynb

import sys
sys.path.append('../src')

from bronze_ingestion import bronze_dag

print("ðŸ§ª Testando Bronze DAG FINAL...")
print("=" * 50)

try:
    result = bronze_dag()
    print("âœ… Bronze DAG: SUCESSO!")
    
    from pathlib import Path
    import pandas as pd
    
    output_file = Path("../data/silver/consumidor_gov_bronze.csv")
    if output_file.exists():
        print(f"\nðŸ“ Arquivo criado: {output_file.name}")
        print(f"ðŸ“Š Tamanho: {output_file.stat().st_size / (1024*1024):.1f} MB")  # Em MB
        
        # Verificar estrutura
        df_check = pd.read_csv(output_file, sep=';', nrows=5)
        print(f"ðŸ“Š Colunas: {len(df_check.columns)}")
        print(f"ðŸ“Š Primeiras colunas: {list(df_check.columns)[:5]}")
        
        # EstatÃ­sticas completas
        df_full = pd.read_csv(output_file, sep=';')
        print(f"\nðŸ“ˆ ESTATÃSTICAS:")
        print(f"   â€¢ Total de registros: {len(df_full):,}")
        print(f"   â€¢ Registros Agibank: {df_full['is_agibank'].sum():,}")
        print(f"   â€¢ Percentual Agibank: {(df_full['is_agibank'].sum()/len(df_full)*100):.2f}%")
        
        # Verificar colunas removidas
        expected_removed = ['Data AnÃ¡lise', 'Data Recusa', 'Prazo Analise Gestor']
        removed_check = [col for col in expected_removed if col not in df_full.columns]
        print(f"   â€¢ Colunas removidas: {len(removed_check)}/{len(expected_removed)}")
        
        print(f"\nðŸŽ‰ Bronze processado com sucesso!")
        print("ðŸ“‹ Arquivo pronto para Silver DAG!")
    
except Exception as e:
    print(f"âŒ Erro: {str(e)}")
    import traceback
    traceback.print_exc()
```
